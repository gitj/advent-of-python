```python
init = np.genfromtxt('input.txt',delimiter=',')
```

```python
ages = init.astype('uint8')
for day in range(80):
    spawn_mask = ages==0
    new_fish = spawn_mask.sum()
    ages = ages - 1
    if new_fish:
        ages[spawn_mask] = 6
        ages = np.concatenate((ages,np.ones((new_fish,))*8)).astype('uint8')
        
    print(day,len(ages))
```

```python
counts = np.bincount(init.astype('int8'))
ages = dict(zip(np.arange(len(counts)),counts))
for day in range(256):
    new_fish = ages.get(0,0)
    if 0 in ages:
        #initially didn't think to delete the old 0 age bin, which can give spurious counts if there is a round with no age 1 fish
        del ages[0]
    for age in range(1,9):
        if age not in ages:
            continue
        ages[age-1] = ages[age]
        del ages[age]
    if new_fish:
        ages[6] = ages.get(6,0) + new_fish
        ages[8] = ages.get(8,0) + new_fish
    print(day,np.array(list(ages.values())).sum(),ages)
```
